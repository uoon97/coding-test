```python
# 13. 섬나라 아일랜드

from collections import deque

def solution():
    N = int(input())
    maps = [list(map(int, input().split())) for _ in range(N)]
    visited = [[0]*N for _ in range(N)]

    dx = [-1, 0, 1, 0, -1, 1, -1, 1]
    dy = [0, -1, 0, 1, -1, -1, 1, 1]

    def bfs(x, y):

        if visited[y][x] == 1 or maps[y][x] == 0:
            return False

        q = deque([(x, y)])
        visited[y][x] = 1

        while q:
            x, y = q.popleft()

            for i in range(8):
                tx = x + dx[i]
                ty = y + dy[i]

                if 0 <= tx < N and 0 <= ty < N:
                    if visited[ty][tx] == 0 and maps[ty][tx] == 1:
                        visited[ty][tx] = 1
                        q.append((tx, ty))

        return True
    
    n_islands = 0

    for i in range(N):
        for j in range(N):
            if bfs(j, i):
                n_islands += 1

    print(n_islands)

solution()
```

```python
# 14. 안전영역
from collections import deque
def solution():
    N = int(input())
    maps = [list(map(int, input().split())) for _ in range(N)]

    def bfs(x, y, height):
        if visited[y][x] == 1 or maps[y][x] <= height:
            return False

        q = deque([(x, y)])
        visited[y][x] = 1

        dx = [-1, 0, 1, 0]
        dy = [0, -1, 0, 1]

        while q:
            cx, cy = q.popleft()

            for i in range(4):
                tx = cx + dx[i]
                ty = cy + dy[i]

                if 0 <= tx < N and 0 <= ty < N:
                    if visited[ty][tx] == 0 and maps[cy][cx] > height:
                        visited[ty][tx] = 1
                        q.append((tx, ty))
        
        return True
    
    def cnt_area(height):
        cnt = 0
        for i in range(N):
            for j in range(N):
                if bfs(j, i, height):
                    cnt += 1
        
        return cnt


    max_height = max(map(max, maps))
    max_cnt = 0
    for height in range(1, max_height + 1):
        visited = [[0]*N for _ in range(N)]
        cnt = cnt_area(height)
        if cnt > max_cnt:
            max_cnt = cnt
    
    return max_cnt

solution()
```

```python
# 15. 토마토
# 시작점이 여러 개 있는 BFS.
from collections import deque
def solution():

    M, N = map(int, input().split())
    matrix = [list(map(int, input().split())) for _ in range(N)]
    starts = [(j, i, 0) for j in range(M) for i in range(N) if matrix[i][j] == 1]

    if all([all(matrix[i]) for i in range(N)]):
        return 0

    visited = [[0]*M for _ in range(N)]
    q = deque(starts)

    dx = [-1, 0, 1, 0]
    dy = [0, -1, 0, 1]

    max_day = 0

    while q:
        x, y, day = q.popleft()

        if day > max_day:
            max_day = day

        for i in range(4):
            tx = x + dx[i]
            ty = y + dy[i]

            if 0 <= tx < M and 0 <= ty < N:
                # 방문 여부를 확인하고, level 단위로 탐색하기 때문에 중복된 부분은 낮은 level로 mapping된다.
                if visited[ty][tx] == 0 and matrix[ty][tx] == 0:
                    visited[ty][tx] = 1
                    q.append((tx, ty, day + 1))

    for i in range(N):
        for j in range(M):
            if visited[i][j] == 0 and matrix[i][j] == 0:
                print(-1)
                exit()
        
    return max_day
```

```python
# 16. 사다리 타기
# 1회 탐색이므로 stack으로 처리해도 됨.
# BFS(queue): append()에서 방문 처리. -> FIFO이기 때문에 문제되지 않음.
# DFS(stack): pop()에서 방문 처리. -> LIFO이기 때문에 문제가 될 수 있음. (방문 처리해놓고 나중에 가버리는 경우)

from collections import deque

# matrix = [list(map(int, input().split())) for _ in range(10)]
start = [(j, i) for j in range(10) for i in range(10) if matrix[i][j] == 2]
line = [i for i in range(10) if matrix[0][i] == 1]

visited = [[0]*10 for _ in range(10)]

s = deque(start)
x, y = start[0]

line_dx = [0, -1, 1]
line_dy = [-1, 0, 0]

step_dx = [-1, 1]
step_dy = [0, 0]


while s:
    x, y = s.pop()

    if y == 0:
        break
    
    visited[y][x] = 1

    if x in line:
        for i in range(3):
            tx = x + line_dx[i]
            ty = y + line_dy[i]

            if 0 <= tx < 10 and 0 <= ty < 10:
                if visited[ty][tx] == 0 and matrix[ty][tx] == 1:
                    s.append((tx, ty))

    
    else:
        for i in range(2):
            tx = x + step_dx[i]
            ty = y + step_dy[i]

            if 0 <= tx < 10 and 0 <= ty < 10:
                if visited[ty][tx] == 0 and matrix[ty][tx] == 1:
                    s.append((tx, ty))

print(x)
```

```python
# 17. 피자 배달 거리
# 각 집으로부터 가장 가까운 피자집: 피자 배달 거리
# 피자집을 M개로 줄였을 때, 피자 배달 거리의 합.
# 순서 상 M개로 줄이는 것이 먼저가 되어야 함.

from itertools import combinations

def solution():
    N, M = map(int, input().split())
    matrix = [list(map(int, input().split())) for _ in range(N)]

    hzs = [(j, i) for j in range(N) for i in range(N) if matrix[i][j] == 1]
    pzs = [(j, i) for j in range(N) for i in range(N) if matrix[i][j] == 2]
    
    min_distance = 999999

    for cpz in combinations(pzs, M):
        distance = 0
        for x, y in hzs:
            distance += min([abs(cx-x) + abs(cy-y) for cx, cy in cpz])

        min_distance = min(min_distance, distance)
            
    return min_distance

solution()
```
