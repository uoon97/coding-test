```python
# 기능 개발 (queue, level 2)

# 앞에가 끝나야 나오니까 무조건 queue
# (5, 10, 1, 1, 20, 1)
# 자기보다 작은 것 popleft() -> while
import math
from collections import deque
def solution(progresses, speeds):
    
    days = deque([math.ceil((100-progresses[i])/speeds[i]) for i in range(len(progresses))])
    res = []
    
    while days:
        curr = days.popleft()
        app = 1
        while days:
            if curr >= days[0]:
                days.popleft()
                app += 1
            
            else:
                break
        
        res.append(app)
    
    return res
```

```python
# 프로세스 (queue, level 2)

from collections import deque

# priorities = [1, 1, 1, 2]
# location = 2

def solution(priorities, location):
    q = deque([(1, 2)])
    
    q = deque(enumerate(priorities))
    cnt = 0
    
    while q:
        loc, v = q.popleft()
        
        if len(q) == 0:
            cnt += 1
            
            if loc == location:
                break
        
        
        if v >= max(list(zip(*q))[-1]):
            cnt += 1

            if loc == location:
                break

        else:
            q.append((loc, v))
            
    return cnt
```

```python
# 다리를 지나는 트럭 (queue, level 2)

# 다리는 트럭이 먼저 진입한 순서로 빠져나오니, Queue
# 한 번(1초)에는 한 트럭만 진입 가능 / 진입이 끝나고, 다리를 모두 지난 트럭을 빼도록.
# t < 20,000
# queue가 꼭 풀 큐에서 시작할 필요는 없다.

from collections import deque
def solution(bridge_length, weight, truck_weights):
    
    # 1, 2초
    q = deque([])
    i, t = 0, 0
    while i < len(truck_weights):
        
        t += 1
        
        if len(q) == 0:
            q.append((t + bridge_length - 1,  truck_weights[i]))
            i += 1
            if t == q[0][0]:
                q.popleft()
            
        else:
            if sum(list(zip(*q))[-1]) + truck_weights[i] <= weight:
                q.append((t + bridge_length - 1,  truck_weights[i]))
                i += 1
            
            if t == q[0][0]:
                q.popleft()
        
    res = q[-1][0] + 1
    
    return res
```

```python
# 조이스틱 (Greedy, level 2)

# min (뒤로 시작할지, 앞으로 시작할지)

# 중복 방문 허용하고, level이 len(name)-1 넘어가는 것은 컷하는 DFS.
# 대신 A가 아닌 것을 카운트하는 것은 중복을 허용하지 않는다.

min_level = 20

def solution(name):
    cnt = 0
    for letter in name:
        cnt += min(ord(letter)-ord('A'), ord('Z')-ord(letter)+1)
    
    def dfs(loc, level, cnt_notA, notA):
        
        if level == len(name)-1:
            return
        
        if cnt_notA == notA:
            global min_level
            min_level = min(level, min_level)
            return
        
        if -len(name) <= (loc-1) < len(name):
            if name[loc-1] != 'A' and visited[loc-1] == 0:
                cnt_notA += 1
                visited[loc-1] = 1
                dfs(loc-1, level+1, cnt_notA, notA)
                visited[loc-1] = 0
                cnt_notA -= 1
            else:
                dfs(loc-1, level+1, cnt_notA, notA)
                
        
        if -len(name) <= (loc+1) < len(name):
            if name[loc+1] != 'A' and visited[loc+1] == 0:
                cnt_notA += 1
                visited[loc+1] = 1
                dfs(loc+1, level+1, cnt_notA, notA)
                visited[loc+1] = 0
                cnt_notA -= 1
            else:
                dfs(loc+1, level+1, cnt_notA, notA)
                
            
    notA = len([letter for letter in name if letter != 'A'])
    visited = [0]*len(name)
    cnt_notA = 0
    
    if name[0] != 'A':
        visited[0] = 1
        cnt_notA += 1
        
    dfs(0, 0, cnt_notA, notA)
    global min_level
    min_level = min(min_level, len(name)-1)
    cnt += min_level
        
    return cnt
```

```python
# 정수 삼각형 (DP, level 3)

def solution(triangle):
    
    dp = [[0]*len(level) for level in triangle]
    
    
    # Bottom up
    for i in range(len(triangle)):
        
        if i == 0:
            dp[0][0] = triangle[0][0]
        
        else:
            for j in range(len(triangle[i])):
                if 0 <= i-1 < len(triangle)-1 and 0 <= j-1 < len(triangle[i-1]): left = dp[i-1][j-1]
                else: left = 0
                if 0 <= i-1 < len(triangle)-1 and 0 <= j < len(triangle[i-1]): right = dp[i-1][j]
                else: right = 0
                
                
                dp[i][j] = max(left, right) + triangle[i][j]
    
    return max(dp[len(triangle)-1])

solution([[7],[3,8],[8,1,0],[2,7,4,4],[4,5,2,6,5]])
```
